```python
#SGD Optimizer Test folder Confusion Matrix
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

binary  = np.array([[124, 16],
                    [14, 126]])

fig, ax = plot_confusion_matrix(conf_mat=binary,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
plt.show()
print("Class Label= 0:Autistic")
print("Class Label= 1:Non_Autistic")
# Extract the true positive, true negative, false positive, and false negative values from the confusion matrix
tp=124
fp=16
fn=14
tn=126

 
# Print the true positive, true negative, false positive, and false negative values
print("True Positive (TP): ", tp)
print("True Negative (TN): ", tn)
print("False Positive (FP): ", fp)
print("False Negative (FN): ", fn)
 
# Calculate accuracy
accuracy = (tp + tn) / (tp + tn + fp + fn)
 
# Calculate precision
precision = tp / (tp + fp)
 
# Calculate recall
recall = tp / (tp + fn)
 
# Calculate F1-score
f1_score = 2 * (precision * recall) / (precision + recall)
# Print the accuracy, precision, recall, and F1-score
print("\n\nMetrics:")
print("Total number of image:280,Autistic:140,Non_Autistic:140")
print("Validation-Test-Accuracy: ", round(accuracy, 2))
print("Precision: ", round(precision, 2))
print("Recall: ", round(recall, 2))
print("F1-score: ", round(f1_score, 2))
 
# Print the formulas for accuracy, precision, recall, and F1-score
print("\n\nFormulas:")
print("Validation-Test-Accuracy: (TP + TN) / (TP + TN + FP + FN)")
print("Precision: TP / (TP + FP)")
print("Recall: TP / (TP + FN)")
print("F1-score: 2 * (Precision * Recall) / (Precision + Recall)")
 
```
